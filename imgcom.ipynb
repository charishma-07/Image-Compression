```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from numpy import asarray
rgbimg=Image.open('ex.jpg')

data=asarray(rgbimg)

data=np.sum(data,axis=2)
img=plt.imshow(data)
img.set_cmap('gray')

print(data.shape)
np.linalg.matrix_rank(data)
```

```python
u,s,v=np.linalg.svd(data,full_matrices = False)
print(u.shape)
print(s.shape)
print(v.T.shape)
```

```python
k=200
comimg=u[:,:k+1]@np.diag(s[:k+1])@v[:k+1,:]
img=plt.imshow(comimg)
img.set_cmap('gray')
```
